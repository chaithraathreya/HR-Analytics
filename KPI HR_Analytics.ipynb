```python
# Trim department values
df['Department'] = df['Department'].str.strip()

#Active employees by department
active_by_dept = (
    df[df['Status'] == 'Active']
    .groupby('Department')['EmployeeID']
    .count()
    .sort_values(ascending=False)
)

active_by_dept
```

```python
#Yearly Employee Intake
df['Hire_Year'] = df['HireDate'].dt.year.astype('Int64')
df['Hire_Year'] = df['Hire_Year'].astype(str).replace('<NA>', 'Unknown')

yearly_hired_count = (
    df.groupby('Hire_Year')['EmployeeID']
    .nunique()
    .reset_index(name='Hired_Count')
    .sort_values('Hire_Year')
)

yearly_hired_count
```

```python
#Gender wise avg_performance rating
df['Gender'] = df['Gender'].str.strip()

avg_perf_gender = (
    df[df['PerformanceRating'].notna()]
    .groupby('Gender')['PerformanceRating']
    .mean()
    .round(2)
    .sort_values(ascending=False)
)

avg_perf_gender
```

```python
# Trim Jobtitle values
df['JobTitle'] = df['JobTitle'].str.strip()

#Active employees by JobTitle
active_by_dept = (
    df[df['Status'] == 'Active']
    .groupby('JobTitle')['EmployeeID']
    .count()
    .sort_values(ascending=False)
)

active_by_dept
```

```python
today = pd.Timestamp.today()

#Average tenure
df['tenure_years'] = (
    (df['TerminationDate'].fillna(today) - df['HireDate'])
    .dt.days / 365
)

#average tenure for all the employees
avg_tenure_all = df['tenure_years'].mean().round(2)
avg_tenure_all
```

```python
#Average tenure in years by gender
avg_tenure_gender = (
    df[df['HireDate'].notna()]
    .groupby('Gender')['tenure_years']
    .mean()
    .round(2)
    .sort_values(ascending=False)
)

avg_tenure_gender
```

```python
# Trim Jobtitle values
df['Department'] = df['Department'].str.strip()

#Active employees by JobTitle
term_by_dept = (
    df[df['Status'] == 'Terminated']
    .groupby('Department')['EmployeeID']
    .count()
    .sort_values(ascending=False)
)

term_by_dept
```

```python
#Yearly Employee Terminated
df['Term_Year'] = df['TerminationDate'].dt.year.astype('Int64')
df['Term_Year'] = df['Term_Year'].astype(str).replace('<NA>', 'Unknown')

yearly_term_count = (
    df.groupby('Term_Year')['EmployeeID']
    .nunique()
    .reset_index(name='Term_Count')
    .sort_values('Term_Year')
)

yearly_term_count
```

```python
# Trim Jobtitle values
df['JobTitle'] = df['JobTitle'].str.strip()

#Terminated  employees by JobTitle
term_by_dept = (
    df[df['Status'] == 'Terminated']
    .groupby('JobTitle')['EmployeeID']
    .count()
    .sort_values(ascending=False)
)

term_by_dept
```

```python
#Average Salary
avg_salary = (
    df[df['Salary'].notna() & (df['Salary'] > 0)]
    ['Salary']
    .mean()
    .round(2)
)

avg_salary
```

```python
#Total employees left
employees_left = df['TerminationDate'].notna().sum()
total_employees = len(df)

#Attrition Rate(%)
attrition_rate = round((employees_left / total_employees) * 100, 2)
attrition_rate
```

```python
#Attrition by Gender
attrition_by_gender = (
    df.groupby('Gender')
    .apply(lambda x: round(
        (x['TerminationDate'].notna().sum() / len(x)) * 100, 2
    ))
    .sort_values(ascending=False)
)

attrition_by_gender
```

```python
#Create Salary Slab
bins = [0, 60000, 80000, 100000, 120000, 140000, 160000, float('inf')]
labels = [
    'Below 60K', '60K-79K', '80K-99K',
    '100K-119K', '120K-139K', '140K-159K', '160K+'
]

df['salary_slab'] = pd.cut(
    df['Salary'],
    bins=bins,
    labels=labels,
    right=False
)

df[['Salary', 'salary_slab']].head()
```

```python
#Attrition rate by salary slab
attrition_by_slab = (
    df[df['salary_slab'].notna()]
    .groupby('salary_slab')
    .agg(
        employee_count=('EmployeeID', 'count'),
        employees_left=('TerminationDate', lambda x: x.notna().sum())
    )
)

attrition_by_slab['attrition_rate_percent'] = (
    attrition_by_slab['employees_left']
    / attrition_by_slab['employee_count'] * 100
).round(2)

attrition_by_slab = attrition_by_slab.sort_index()

attrition_by_slab
```

```python
#Attrition rate by gender
attrition_by_gender = (
    df[df['Gender'].notna()]
    .groupby('Gender')
    .agg(
        employee_count=('EmployeeID', 'count'),
        employees_left=('TerminationDate', lambda x: x.notna().sum())
    )
)

attrition_by_gender['attrition_rate_percent'] = (
    attrition_by_gender['employees_left']
    / attrition_by_gender['employee_count'] * 100
).round(2)

attrition_by_gender.sort_values(
    by='attrition_rate_percent',
    ascending=False
)
```

```python
avg_salary_slab_gender = (
    df[df['Salary'].notna()]
    .groupby(['Gender', 'salary_slab'])
    .agg(
        employee_count=('EmployeeID', 'count'),
        avg_salary=('Salary', 'mean')
    )
    .round(2)
    .sort_index()
)

avg_salary_slab_gender

```
